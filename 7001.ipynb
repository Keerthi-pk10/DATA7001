```python
import pandas as pd
from scipy.stats import kendalltau

# Assuming 'df' is your DataFrame with relevant numerical columns
columns = df.columns

# Dictionary to store results
kendall_results = {}

for i in range(len(columns)):
    for j in range(i + 1, len(columns)):
        # Perform Kendall's Tau correlation test
        tau, p_value = kendalltau(df[columns[i]], df[columns[j]])

        # Store results with the variable pair as key
        kendall_results[f'{columns[i]} vs {columns[j]}'] = (tau, p_value)

# Convert the results into a DataFrame for better visualization
kendall_df = pd.DataFrame(kendall_results, index=['Kendall\'s Tau', 'p-value']).T

# Interpretation Function
def interpret_kendall_result(tau, p_value):
    # Interpretation of the correlation strength
    if abs(tau) < 0.1:
        strength = "very weak"
    elif abs(tau) < 0.3:
        strength = "weak"
    elif abs(tau) < 0.5:
        strength = "moderate"
    elif abs(tau) < 0.7:
        strength = "strong"
    else:
        strength = "very strong"

    # Determine statistical significance
    if p_value < 0.05:
        significance = "statistically significant"
    else:
        significance = "not statistically significant"

    return f"The relationship is {strength} and {significance}."

# Apply the interpretation to each pair
kendall_df['Interpretation'] = kendall_df.apply(lambda row: interpret_kendall_result(row['Kendall\'s Tau'], row['p-value']), axis=1)

# Filter for moderate, strong, and very strong correlations that are statistically significant
def filter_significant_strong_moderate(row):
    tau = row['Kendall\'s Tau']
    p_value = row['p-value']
    # Filter for moderate (0.3 to 0.5), strong (0.5 to 0.7), and very strong (>0.7) correlations
    if p_value < 0.05 and abs(tau) >= 0.3:
        return True
    return False

# Apply the filter
filtered_kendall_df = kendall_df[kendall_df.apply(filter_significant_strong_moderate, axis=1)]

# Set display options for Pandas
pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.width', None)  # Adjust width to display everything
pd.set_option('display.max_colwidth', None)  # Show full content of each column

# Display the filtered results
print(filtered_kendall_df)
```

```python
# Required Libraries
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Independent and dependent variables
# Independent and dependent variables
# Define X (features) and y (target)
X = df_numeric.drop(columns=['QLD_RRP'])  # Features (drop the target column)
y = df_numeric['QLD_RRP']  # Target        # Dependent variable

# Feature scaling (important for SVR)
scaler_X = StandardScaler()
scaler_y = StandardScaler()

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y.values.reshape(-1, 1)).ravel()  # SVR requires 1D array for y
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)

# SVM Model (using RBF kernel for non-linear regression)
svr_model = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=0.1)
svr_model.fit(X_train, y_train)

# Predictions
y_train_pred_svr = svr_model.predict(X_train)
y_test_pred_svr = svr_model.predict(X_test)

# Inverse transform to get predictions back to original scale
# Reshape the predictions to 2D arrays
y_train_pred_svr_orig = scaler_y.inverse_transform(y_train_pred_svr.reshape(-1, 1))
y_test_pred_svr_orig = scaler_y.inverse_transform(y_test_pred_svr.reshape(-1, 1))
y_train_orig = scaler_y.inverse_transform(y_train.reshape(-1, 1))
y_test_orig = scaler_y.inverse_transform(y_test.reshape(-1, 1))

# Evaluate the model (MSE and R-squared)
train_mse_svr = mean_squared_error(y_train_orig, y_train_pred_svr_orig)
test_mse_svr = mean_squared_error(y_test_orig, y_test_pred_svr_orig)
train_r2_svr = r2_score(y_train_orig, y_train_pred_svr_orig)
test_r2_svr = r2_score(y_test_orig, y_test_pred_svr_orig)

print(f"Train MSE (SVM): {train_mse_svr}, Train R2: {train_r2_svr}")
print(f"Test MSE (SVM): {test_mse_svr}, Test R2: {test_r2_svr}")
```
